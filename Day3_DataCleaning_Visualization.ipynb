```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
    'age': [25, 30, 35, None, 40, 120, 28, None, 32],
    'salary': [30000, 50000, None, 45000, 52000, None, 38000, 70000, 60000],
    'city': ['Delhi', 'Mumbai', 'Delhi', 'Chennai', None, 'Delhi', 'Mumbai', 'Delhi', 'Chennai']
}

df = pd.DataFrame(data)
df
```

```python
df.isnull().sum()
```

```python
df['age'].fillna(df['age'].mean(), inplace=True)
df['salary'].fillna(df['salary'].mean(), inplace=True)
df['city'].fillna(df['city'].mode()[0], inplace=True)

df
```

```python
sns.boxplot(df['age'])
plt.show()
```

```python
df = df[df['age'] < 100]
df
```

```python
df.describe()
```

```python
sns.histplot(df['age'], kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
sns.scatterplot(x=df['age'], y=df['salary'])
plt.title("Age vs Salary")
plt.show()
```

```python
sns.countplot(x=df['city'])
plt.title("City Count")
plt.show()
```
