```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
    'age': [25, 30, 35, None, 40, 120, 28, None, 32],
    'salary': [30000, 50000, None, 45000, 52000, None, 38000, 70000, 60000],
    'city': ['Delhi', 'Mumbai', 'Delhi', 'Chennai', None, 'Delhi', 'Mumbai', 'Delhi', 'Chennai']
}

df = pd.DataFrame(data)
df
```

```python
corr = df.corr(numeric_only=True)
corr
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.scatter(df['age'], df['salary'])
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs Salary Scatter Plot")
plt.show()
```

```python
sns.countplot(x='city', data=df)
plt.title("City Count Plot")
plt.show()
```
